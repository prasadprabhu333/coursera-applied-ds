```python
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import matplotlib.animation as animation
import numpy as np

%matplotlib notebook

n = 100
# generate 4 random variables from the random, gamma, exponential, and uniform distributions
normal_data = np.random.normal(5, 1, 10000)
gamma_data = np.random.gamma(2, 1, 10000)
exp_data = np.random.exponential(1, 10000)
uniform_data = np.random.uniform(1,10, 10000)

# common variables
bins = np.arange(1, 10, 0.5)
fig = plt.figure(figsize=(10,8))


def update(curr):
    if curr == n:
        a.event_source.stop()
    plt.cla()
    gspec = gridspec.GridSpec(2, 2)
    sample_size = curr * 100
    normal_hist = plt.subplot(gspec[0, 0], title=f'Normal Distribution, n={sample_size}', 
                              ylim=(0, 1.0))
    gamma_hist = plt.subplot(gspec[0, 1], title=f'Gamma Distribution, n={sample_size}',
                            ylim=(0, 1.0))
    exp_hist = plt.subplot(gspec[1,0], title=f'Exponential Distribution, n={sample_size}',
                            ylim=(0, 1.0))
    uniform_hist = plt.subplot(gspec[1, 1], title=f'Uniform Distribution, n={sample_size}',
                            ylim=(0, 1.0))
    
    _n, _bins, rectangles1 = normal_hist.hist(normal_data[:sample_size], 
                                              density=True, 
                                              bins=bins, 
                                              color='blue', 
                                              alpha=0.5)
    _n, _bins, rectangles2 = gamma_hist.hist(gamma_data[:sample_size], 
                                             density=True, 
                                             bins=bins, 
                                             color='orange', 
                                             alpha=0.5)
    _n, _bins, rectangles3 = exp_hist.hist(exp_data[:sample_size], 
                                           density=True, 
                                           bins=bins, 
                                           color='lightblue', 
                                           alpha=0.5)
    _n, _bins, rectangles4 = uniform_hist.hist(uniform_data[:sample_size], 
                                               density=True, 
                                               bins=bins, 
                                               alpha=0.5)    

    return rectangles1 + rectangles2 + rectangles3 + rectangles4


a = animation.FuncAnimation(fig, update, interval = 100)
fig.show()
```
